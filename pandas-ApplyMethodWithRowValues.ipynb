```python
import pandas as pd
bond=pd.read_csv('jamesbond.csv',index_col='Film')
bond.sort_values(by="Film", inplace=True)
bond.head()
```

```python
def good_movie(row):
    actor=row[1]
    budget=row[4]
    
    if actor=="Pierce Brosnan":
        return "The Best"
    elif actor=='Sean Connery' and budget >40:
        return "Enjoyable"
    else:
        return "I have no idea"
    
bond.apply(good_movie,axis=1)
```
